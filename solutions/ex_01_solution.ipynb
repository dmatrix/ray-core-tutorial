```python
import math
import numpy as np
import logging
from typing import List
import ray
```

```python
@ray.remote
def compute_intensive_func(num: int, dims=10) -> List[np.array]:
    dot_products = []
    for _ in range(num):
        # Create a dims x dims matrix
        x = np.random.rand(dims, dims)
        y = np.random.rand(dims, dims)
        
        # Create a dot product of itself
        dot_products.append(np.dot(x, y))
    return dot_products
```

```python
if ray.is_initialized:
    ray.shutdown()
ray.init(logging_level=logging.ERROR)
```

```python
%%time
results = [compute_intensive_func.remote(i, 5_000) for i in range(5)]
```
